```python
# Import Libraries
import numpy as np
import pandas as pd
```

```python
# Load Datasets

# Economic Data
economic_real = pd.read_csv('economic_variables.csv')
economic_nominal = pd.read_csv('economic_nominal_variables.csv')

# Climate Data
climate_aggregated = pd.read_csv('climate_aggregated_variables.csv')
climate_midas = pd.read_csv('climate_midas_variables.csv')
```

## Missing Values

#### Economic Data

```python
# Assess Missing Values in Economic Data
economic_real.rename(columns={'geo': 'NUTS_ID'}, inplace=True)
economic_real = economic_real.sort_values(by=['NUTS_ID', 'year']).reset_index(drop=True)
print(economic_real.info())
print(' ')
economic_nominal.rename(columns={'geo': 'NUTS_ID'}, inplace=True)
economic_nominal = economic_nominal.sort_values(by=['NUTS_ID', 'year']).reset_index(drop=True)
print(economic_nominal.info())
```

```python
# Fill Missing Values using Previous Years where Possible

# Real Data
for i in range(1, len(economic_real)):
    if economic_real.loc[i, 'NUTS_ID'] == economic_real.loc[i-1, 'NUTS_ID']:
        for j in economic_real.columns.tolist()[3:]:
            if economic_real.loc[i, j] == np.nan:
                economic_real.loc[i, j] = economic_real.loc[i-1, j]
                
# Nominal Data
for i in range(1, len(economic_nominal)):
    if economic_nominal.loc[i, 'NUTS_ID'] == economic_nominal.loc[i-1, 'NUTS_ID']:
        for j in economic_nominal.columns.tolist()[3:]:
            if economic_nominal.loc[i, j] == np.nan:
                economic_nominal.loc[i, j] = economic_nominal.loc[i-1, j]
```

#### Climate Data

```python
# View Missing Values in Climate Data
climate_aggregated = climate_aggregated.sort_values(by=['NUTS_ID', 'year']).reset_index(drop=True)
print(climate_aggregated.columns())
print(' ')
climate_midas = climate_midas.sort_values(by=['NUTS_ID', 'year']).reset_index(drop=True)
print(climate_midas.columns())
```
